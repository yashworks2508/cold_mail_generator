```python

from langchain_community.document_loaders import WebBaseLoader
from bs4 import BeautifulSoup
loader = WebBaseLoader("https://www.google.com/about/careers/applications/jobs/results/87524772086194886-senior-system-software-engineer-embedded-pixel-watch?category=DATA_CENTER_OPERATIONS&category=DEVELOPER_RELATIONS&category=HARDWARE_ENGINEERING&category=INFORMATION_TECHNOLOGY&category=MANUFACTURING_SUPPLY_CHAIN&category=NETWORK_ENGINEERING&category=PRODUCT_MANAGEMENT&category=PROGRAM_MANAGEMENT&category=SOFTWARE_ENGINEERING&category=TECHNICAL_INFRASTRUCTURE_ENGINEERING&category=TECHNICAL_SOLUTIONS&category=TECHNICAL_WRITING&q=")
page_data= loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate
prompt_extract = PromptTemplate.from_template("""
### SCRAPED TEXT FROM WEBSITE: {page_data}
### INSTRUCTION:
The scraped text is from the career page. Your job is to extract the job postings with the following keys: role, experience.
Only return the valid JSON.
#### VALID JSON (NO PREAMBLE):
""")


chain_extract = prompt_extract | llm


extracted = chain_extract.invoke({"page_data": page_data})
print(extracted.content)
```

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\lachi\Documents\my_portfolio.csv")
df
```

```python
import chromadb
import uuid
import pandas as pd
from chromadb import PersistentClient

client = PersistentClient('vectorstore')
collection = client.get_or_create_collection(name="portfolio")
if collection.count() == 0:
    for _, row in df.iterrows():
        collection.add(
            documents=[row['Techstack']],   
            metadatas=[{"links": row['Links']}],  
            ids=[str(uuid.uuid4())]          
        )
```

```python
links = collection.query(
    query_texts=["Python", "React"],
    n_results=1
)
links
```

```python
prompt_email = PromptTemplate.from_template(
"""
### JOB DESCRIPTION:
{job_description}

### INSTRUCTION:
You are Yash, business development executive at Atliq. Atliq is an AI & Soft-
ware company that helps businesses increase efficiency & automate processes through tailored solutions.
Over our experience, we have empowered numerous enterprises with tailored solu-
tions, improving productivity and efficiency overall.
Your job is to write a cold email to the client regarding the job mentioned above.
Also add the most relevant ones from the internet.
Remember you are Yash, BDE at Atliq.
Do not provide any preamble.
## EMAIL (NO PREAMBLE):
"""
)
chain_email = prompt_email | llm

email_content = chain_email.invoke({"job_description": str(extracted), "link_list": links})
print(email_content.content)
```
